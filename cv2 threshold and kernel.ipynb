```python
import numpy as np
```

```python
k = np.ones( (3,3 ))  / 25
```

```python
k
```

```python
k1 = np.array([
      [0, -1, 0], 
      [-1,5,-1], 
      [0,-1,0]
])
```

```python
k1
```

```python
photo
```

```python
nphoto = cv2.filter2D(src=photo, ddepth=-1,  kernel=k1 )
```

```python
nphoto
```

```python
cv2.imshow("hi", photo)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(1)
```

```python
while True:
    ret, photo = cap.read()
    nphoto = cv2.filter2D(src=photo, ddepth=-1,  kernel=k1 )
    fphoto = cv2.cvtColor(nphoto, cv2.COLOR_BGR2GRAY)
    cv2.imshow("hi", fphoto)
    if cv2.waitKey(100) == 13:
        break
        
cv2.destroyAllWindows()
```

```python
bphoto = cv2.blur(photo, (10,10)  )
```

```python
cv2.imshow("hi", bphoto)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
nphoto = cv2.medianBlur(src=photo, ksize=5 )
```

```python
cv2.imshow("hi", nphoto)
cv2.waitKey()
cv2.destroyAllWindows()
```
